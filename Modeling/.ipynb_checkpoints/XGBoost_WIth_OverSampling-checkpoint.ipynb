```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Read the datasets from the CSV files
x_training = pd.read_csv("data_split/X_resampled_train.csv")
y_training = pd.read_csv('data_split/y_resampled_train.csv').iloc[:, 0]

x_validation = pd.read_csv('data_split/X_val.csv')
y_validation = pd.read_csv('data_split/y_val.csv').iloc[:, 0]


x_test = pd.read_csv('data_split/X_test.csv')
y_test = pd.read_csv('data_split/y_test.csv').iloc[:, 0]
```

```python


# Convert data to DMatrix format
import xgboost


dtrain = xgboost.DMatrix(x_training, label=y_training)
dval = xgboost.DMatrix(x_validation, label=y_validation)

# Define the parameters for tuning
parameters = {
    'max_depth': [3, 6, 9],
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [100, 500, 1000]
}

# Create an XGBoost classifier
xgb_model = xgboost.XGBClassifier()

# Create a GridSearchCV object
grid_search = GridSearchCV(estimator=xgb_model, param_grid=parameters, scoring='accuracy', cv=3, n_jobs=-1)

# Fit the model with the parameter candidates
grid_search.fit(x_training, y_training)

# Print best parameters
print("Best Parameters: ", grid_search.best_params_)

# Use the best model to predict on the testing set
y_test_pred = grid_search.predict(x_test)

# Evaluate the model on the testing set
accuracy = accuracy_score(y_test, y_test_pred)
print(f'Accuracy on the testing set: {accuracy}')
```

```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, cohen_kappa_score, roc_auc_score


# Create an XGBoost DMatrix for training and validation data
dtrain = xgb.DMatrix(x_training, label=y_training)
dval = xgb.DMatrix(x_validation, label=y_validation)

# Define the parameters for tuning
parameters = {
    'max_depth': [3, 6, 9],
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [100, 500, 1000]
}

# Create an XGBoost classifier
xgb_model = xgb.XGBClassifier(objective='multi:softmax', num_class=6, multi_class='ovr')

# Create a GridSearchCV object
grid_search = GridSearchCV(estimator=xgb_model, param_grid=parameters, scoring='accuracy', cv=3, n_jobs=-1)

# Fit the model with the parameter candidates
grid_search.fit(x_training, y_training)

# Print best parameters
print("Best Parameters: ", grid_search.best_params_)

# Use the best model to predict on the testing set
y_test_pred = grid_search.predict(x_test)
y_train_pred = grid_search.predict(x_training)


# Compute metrics
test_accuracy = accuracy_score(y_test, y_test_pred)
train_accuracy = accuracy_score(y_training, y_train_pred)

precision = precision_score(y_test, y_test_pred, average='micro')
recall = recall_score(y_test, y_test_pred, average='micro')
f1 = f1_score(y_test, y_test_pred, average='micro')

classification_rep = classification_report(y_test, y_test_pred)

# Compute ROC AUC score (one-vs-rest)
roc_auc = 0
for i in range(6):  # Assuming there are 3 classes
    roc_auc += roc_auc_score(y_test == i, y_test_pred == i)
roc_auc /= 6

kappa = cohen_kappa_score(y_test, y_test_pred)

# Print the metrics
print(f'Test Accuracy: {test_accuracy}')
print(f'Training Accuracy: {train_accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
print(f'Classification Report:\n{classification_rep}')
print(f'ROC AUC Score: {roc_auc}')
print(f"Cohen's Kappa Score: {kappa}")
```
